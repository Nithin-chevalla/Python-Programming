# While loop

> if we want to execute a group of statements iteratively until some condition false, then go for while loop.

```python
#to print number from 1 to 10 while loop
x=1
while x<=10:
    print(x)
    x=x+1
```

```python
#display the sum of first n number
n=int(input("enter a Number"))
sum=0
i=1
while i<=n:
    sum=sum+i
    i=i+1
print("the sum of first ",n,"number is",sum)
```

# Nested loop

loop within the loop

```python
for i in range(4):
    for j in range(4):
        print("i=",i,"j=",j)
```

# Tranfer Statements

> Break

- we can use break statement inside loops to break loop execution based on some condition.

```python
for i in range(10):
    if i==7:
        print("processing is enough,please break")
        break
    print(i)
```

```python
cart=[100,200,300,400,500,600,700,800,900,1000]
for item in cart:
    if item>500:
        print("you will 20% discount on this")
        break
    print(item)
```

# Continue

- we can use continue statement to skip current iteration and continue to next.

```python
for i in range(10):
    if i%2==0:
        continue #it will skip when it is true or correct
    print(i)
```

```python
cart=[100,200,300,400,500,600,700,800,900,1000]
for item in cart:
    if item>=500:
        print("we cannot process this item",item)
        continue
    print(item)
```
