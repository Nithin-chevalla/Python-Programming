```python
b=[10,20,30,40,60]
```

```python
type(b)
```

```python
x=bytes(b)
```

```python
type(x)
```

# Bytesarray Data Types

It is same as Bytes, But it is mutable

```python
y=[11,12,13,14,15]
```

```python
ba=bytearray(y)
```

```python
for i in y:
    print(i)
```

```python
#checking the value of the index in Bytearray
ba[0]
```

```python
# changing or replacing the value in bytearray based on index
ba[0]=10
```

- 2. heterogenous elements are allowed
- 3. duplicates are allowed
- 4. growable in nature
- 5. It is mutable
