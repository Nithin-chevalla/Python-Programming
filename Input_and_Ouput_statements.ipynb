### How to take multiple input at a time

```python
a,b=[int(x) for x in input("enter 2 numbers:").split()] #using split function to take multiple inputs
print("product of this:",a*b)
```

```python
p,q,r=[float(x) for x in input("enter 3 numbers:").split()]
print("product of this:",p*q*r)
```

```python
#splitting using " , "or comma
p,q,r=[float(x) for x in input("enter 3 numbers:").split(",")]
print("product of this:",p*q*r)
```

# formatted string

```python
# Method 1
name="pavan"
salary=25000
friend="sunny"
print("hello {0} your salary is {1} and friend is {2}".format(name,salary,friend))
```

```python
# Method 2
name="pavan"
salary=25000
friend="sunny"
print("hello {x} your salary is {y} and friend is {z}".format(x=name,y=salary,z=friend))
```
