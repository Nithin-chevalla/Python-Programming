# Removing the spaces from the string

- There are 3 methods to remove spaces
1. rstrip() ===> remove space right hand side
2. lstrip() ===> remove space left hand side
3. strip() ===> remove space both sides

```python
city=input("enter your city")
scity=city.strip()
if scity=="Hyderabad":
    print("hello hyderabadi...Adaab")
elif scity=="Chennai":
    print("hello madrasi...Vanakkam")
elif scity=="bengaluru":
    print("hello subhodya")
else:
    print("Enter a valid city name")
```

# Finding sub-string

we can find out based on direction 
1. Forward Direction
- find() ==> Return first occurance of substring
- index() ==> return substring 

2. Backward direction 
- rfind()
- rindex()

```python
s="Machine learning is learning machine"
```

```python
s.find("machine") # returing the small letter machine
```

```python
s.find("Machine") # returning the capital letter Machine
```

```python
s.find('a',0,6) # finding the char from 0 to 6 index
```

```python
s.find('a',3,20) # finding the char from 3 to 15 index
```
