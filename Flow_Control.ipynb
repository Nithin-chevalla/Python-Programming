# Flow Control
- It describes the order in which the statements will be executed.
- There are 3 types:
- 1. Iterative Statements - for,while
  2. Transfer Statements - break,continue,pass
  3. Conditional Statements - if, else, elif

# Conditional Statements

* IF Statement
  - ### Single statement-  if condition:statement
  - ### Multiple Statements- if condition:
                                    statement-1
                                    statement-2
                                    statement-3

```python
name=input("enter name")
if name=="vinay":
    print("hii",name)
```

```python
if name=="vinay":
    print("hello",name)
    print(name+name)
    print("Hii",name,"how are you?")
```

```python
x=int(input("enter a number"))
if x==0:
    print("enter valid number")
elif x<=100:
    print("it is in between 1 to 100")
else:
    print("more than 100")
```

```python
x=int(input("enter first number"))
y=int(input("enter second number"))
z=int(input("enter third number"))
if x<y and x<z:
    print("smallest number is",x)
elif y<x and y<z:
    print("smallest number is",y)
elif z<y and z<x:
    print("smallest number is",z)
else:
    print("enter a valid number")
```

```python
#method 2
x,y,z=[int(a) for a in input("enter 3 numbers").split()]
if x<y and x<z:
    print("smallest Number is",x)
elif y<x and y<z:
    print("smallest number is",y)
elif z<x and z<y:
    print("smallest number is",z)
else:
    print("enter valid number")
```
