```python
# A name in python program is called identifier.
# It can be the name of variable name,function name ,module name or class name.
```

### Rule sto define Identifiers in Python

```python
# Rule-1 Only Alphabets symbols are allowed ( UPPER and lower ) A-Z ,a-z.
# Rule-2 0-9 digits are allowed.
# Rule-3 Special symbols like ( _ ) underscore Symbols are allowed.
# Rule-4 $ symbol is not allowed.
# Rule-5 Identifier should with character not with number.
# Rule-6 Identifier are case sensitive.
# Rule-7 We can't use Reserve words as Identifiers
```

```python
# Example rule-4
cash=20
ca$h=20
```

```python
# Example rule-5
number=30
number123=500
1number=20
```

```python
# Example Rule-6
total=20
TOTAL=30
Total=40
# All are different

```

```python
# Example Rule-7
import keyword
keyword.kwlist
```
