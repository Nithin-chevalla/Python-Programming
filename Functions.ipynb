# Variable length arguments

sometimes we pass variable number of arguments

```python
# def f1(*n):  ==> syntax
```

```python
def sum(*n):
    total=0
    for i in n:
        total=total+i
    print(total)
```

```python
sum(10)
```

```python
sum(10,20,30,40,50,60)
```

```python
sum(10,20)
sum(30,40)
sum(40,50)
```

```python
# write a function for factorial
```

```python
def fact(n):
    result=1
    for i in range(1,n+1): #using range function
        result=result*i
    return result
```

```python
fact(5)
```

```python
def fact1(n):
    if n==0:
        result=1
    else:
        result=n*fact1(n-1) # recursion function - which can call itself
    return result
```

```python
fact1(5)
```

# filter()

- we can use filter() to filter values from the given sequence based on some condition

filter(function,sequence) ==>syntax

```python
# to find the even numbers from list using filter()
```

```python
def is_even(x):
    if x%2==0:
        return True
    else:
        return False
```

```python
l=[0,4,3,5,6,78,9,4,2,4,5,67,8,95,]
l1=list(filter(is_even,l))
```

```python
print(l1)
```

```python
# with lambda function
l=[1,2,3,45,24,34,5,36,63,998]
l1=list(filter(lambda x:x%2==0,l))
print(l1)
```

```python
# if we need odd from the above
l=[1,2,3,45,24,34,5,36,63,998]
l1=list(filter(lambda x:x%2!=0,l))
print(l1)
```

# map()

- for every element present in the given sequence , apply some functionality and generate new element with the required modification fo this map()
