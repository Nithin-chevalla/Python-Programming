#### Five Fundamental Data type In Python



* 1. Integer int=(number (-ve,0,+ve))

* 2. Float float=( Decimal Number)

* 3. String str=( characters, alphabets )

* 4. Complex = real + Imaginary= 4+7j

* 5. Boolean = True/False

#### Composite Data types


- 1. Bytes. 1Byte=8Bits ( 0-255(256) )
- 2. Bytearray. ( Format will change )
- 3. Range. ( Length of number or Range of numbers)
- 4. List.=(Mutuable )
- 5. tuple=(Immutable)
- 6. set
- 7. Frozenset
- 8. dictionary
- 9. None
