# Random module

- It is used to generate random numbers.
- It is having multiple functions, 

# random() function

```python
from random import *
for i in range(10):
    print(random()) # printing random numbers in range
```

# randint() 
- ==>to generate random integer between two given numbers

```python
from random import *
```

```python
for i in range(10):
    print(randint(1,20))
```

```python
dir(random) # dir(module_name) - to check all the members of module
```

# uniform() 
- => it returns random float value between 2 numbers

```python
from random import *
for i in range(10):
    print(uniform(1,10))
```

# randrange([start,stop,step])

```python
import random
for i in range(12):
    print(randrange(1,7,2))
```

# choice() 
- it will not return random numbers
- whereas it returns random objects from the given list/tuple

```python
from random import *
list1=['bunny','sunny','funny','pinny','ginny']
for i in range(10):
    print(choice(list1))
```
