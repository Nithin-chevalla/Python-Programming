```python
#Taking input from Keyboard and print minimum value 
x=int(input("enter the First Number"))
y=int(input("enter the second number"))
z=int(input("enter the Thrid number"))
min= x if x<y and x<z else y if y<x and y<z else z # condition for minimium
min #printing or calling
```

```python
#Taking input from Keyboard and print maximum value 
p=int(input("enter the First Number"))
q=int(input("enter the second number"))
r=int(input("enter the Thrid number"))
max= p if p>q and p>r else q if q>p and q>r else r #condition for max
max #printing or calling
```
