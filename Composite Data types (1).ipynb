# Range Data Type

- This will represent a sequence of numbers.
- The elements which will be presented they will not be modifiable
- i.e Immutable

```python
# range(stop)
a=range(10)
```

```python
a
```

```python
for i in a:
    print(i)
```

```python
# range(start,stop) 
b=range(10,20)
```

```python
b
```

```python
for i in b:
    print(i)
```

```python
# range(start,stop,step)
c=range(2,21,2)
```

```python
for i in c:
    print(i)
```

```python
c[4]
```

```python
# item assignment is not possible in range , we can only print or see the element based on index
```

```python
# generting a list using the range
l=list(range(5,51,5))
```

```python
l
```

```python
# we can generate a tuple using range also
tu=tuple(range(10,101,10))
```

```python
tu
```

# Set Data Type 

- when we want to represent a group of values without duplicates.
- where order is not important, then we should go for set data type.

- 1. Insertion order is not preserved.
  2. Duplicates are not allowed.
  3. Hetrogenous objects are allowed.
  4. index concept is not applicable in set.
  5. it is mutable.
  6. Growable is nature.
  7. we enclosed in {}.

```python
s={10,23.4,5+9j,"Nithin",True,10,20}
```

```python
s
```

```python
# Indexing is not possible
s[0]
```

```python
# it is Growable in nature
s.add(500)
```

```python
s
```

# Frozenset Data Type

- It is same as set only except , Frozenset is immutable
- add() and remove() will not work

```python
s
```

```python
fs=frozenset(s)
```

```python
type(fs)
```

```python
for i in fs:
    print(i)
```

```python
# we cannot add or remove
fs.add(40)
```

```python
fs.remove(20)
```
